# Introduction au Hierarchical clustering(regroupement hiérarchique) 

Les algorithmes de regroupement hiérarchique construisent un hiérarchie de clusters où chaque nœud est une cluster composée des grappes de ses nœuds filles.

Le clustering tente de trouver une structure dans les données en créant des groupes de données ayant des caractéristiques similaires. L'algorithme de clustering le plus connu est probablement K_means, mais il existe un grand nombre de façons de regrouper des observations. Le regroupement hiérarchique est une autre classe d'algorithmes de regroupement qui produit de 1 à n groupes, où n est le nombre d'observations dans l'ensemble de données. En descendant dans la hiérarchie de 1 cluster(contenant toutes les données) à n cluster (chaque observation est sa propre cluster), les clusters deviennent de plus en plus similaires (presque toujours).

Les stratégies de regroupement hiérarchique sont généralement de deux types : Divisives(divisive) et Agglomératives(Agglomerative).
La stratégie de division est descendante, donc vous commencez avec toutes les observations d'un grand groupe et vous les séparez en petits morceaux. Pensez à diviser comme à "diviser" le groupe.
Agglomérat est l'opposé de diviseur, c'est donc ascendant, où chaque observation commence dans sa propre cluster et des paires de cluster sont fusionnées au fur et à mesure qu'elles montent dans la hiérarchie.
L'agglomération signifie rassembler des choses, ce qui est exactement ce que cela fait avec le cluster.
L'approche agglomérative est plus populaire parmi les data scientist.

Autres détails à propos les 2 stratégies de regroupement hiérarchique : 
1) Divisive : 
Le regroupement hiérarchique par division fonctionne en commençant par un groupe contenant l'ensemble des données. L'observation présentant la plus grande dissimilitude moyenne (la plus éloignée de la cluster par une mesure quelconque) est réaffectée à sa propre cluster. Toute observation de l'ancienne cluster plus proche de la nouvelle est assignée à la nouvelle cluster. Ce processus se répète avec la plus grande cluster jusqu'à ce que chaque observation soit sa propre grappe.
2) Agglomerative: 
L'agglomération commence par chaque observation comme à sa propre cluster. Les deux clusters les plus proches sont réunies en une seule cluster. Les deux groupes les plus proches suivantes sont regroupées et ce processus se poursuit jusqu'à ce qu'il n'y ait plus qu'une seule groupe contenant l'ensemble des données.

En général, cela dépend entièrement du type de données, de leur dimensionnalité et, surtout, de leur nature, la connaissance du domaine de l'ensemble de données. En fait, les différentes approches pour définir la distance entre les clusters, distinguer les différents algorithmes.
Comme vous pouvez l'imaginer, il existe de multiples façons de le faire.
La première est appelée "Single-Linkage Clustering". La liaison simple est définie comme la plus courte distance entre 2 points dans chaque groupe, comme les points "a" et "b".(image ci dessus)
La prochaine étape est le regroupement des liens complets"Complete-Linkage Clustering". Cette fois, nous trouvons la plus longue distance entre les points de chaque groupe, comme la distance entre les points "a" et "b".
Le troisième type de lien est le regroupement de liens moyens"Average Linkage Clustering", ou la distance moyenne.
Cela signifie que nous examinons la distance moyenne de chaque point d'un groupe à chaque point d'un autre groupe. Le dernier type de lien à examiner est "Centroid Linkage Clustering". Centroid est la moyenne des ensembles de caractéristiques de points dans un cluster. Cette liaison prend en compte le centroïde de chaque groupe lors de la détermination de la distance minimale.




## Avantages et inconvénients  

# Avantages : 

L'utilisation du regroupement hiérarchique présente trois grands avantages : 
Premièrement, nous n'avons pas besoin de spécifier le nombre de clusters requis pour l'algorithme.
Deuxièmement, la mise en clusters hiérarchiques est facile à mettre en œuvre.
Et troisièmement, le dendrogramme produit est très utile pour comprendre les données.

# inconvénients : 

Il y a aussi quelques inconvénients. Tout d'abord, l'algorithme ne peut jamais annuler les étapes.
Deuxièmement, la complexité du temps pour le regroupement peut entraîner des temps de calcul très longs, par rapport à des algorithmes efficaces, tels que k-Means. Enfin, si nous disposons d'un vaste ensemble de données, il peut devienne difficiles à déterminer le nombre correct de clusters par le dendrogramme.

# Comparasion entre K_means et le regroupement hiérarchique : 

## K_Means : 
1) Beaucoup plus efficace pour les larges datasets.

2) Exige que le nombre de clusters soit spécialisé. 

3) Ne donne qu'un seul partitionnement des données en fonction du nombre prédéfini de clusters.

4) Retourne potentiellement des clusters différents à chaque fois qu'il est exécuté en raison de l'initialisation aléatoire des centroïdes.

# le regroupement hiérarchique : 
1) peut être lent pour les large datasets.

2) Ne nécessite pas le nombre de clusters pour fonctionner.

3) Donne plus d'un partitionnement en fonction de la résolution.

4) Toujours générer les mêmes clusters.
